# Model building for casual users

```python
data.columns
```

```python
df=data[['season', 'yr', 'mnth', 'hr', 'holiday', 'weekday',
       'workingday', 'weathersit', 'atemp', 'hum', 'windspeed']].copy()
```

```python
target=data.casual
```

```python
sns.distplot(target)
```

```python
from sklearn.cross_validation import  train_test_split
X_train, X_test, y_train, y_test=train_test_split(df,target, test_size=.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression(fit_intercept=True,normalize=True,n_jobs=2)
```

```python
lr.fit(X_train,y_train)
```

```python
lr.score(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
# Poission
import numpy as np
y_tr=np.log1p(y_train)
y_ts=np.log1p(y_test)
```

```python
lr=LinearRegression(fit_intercept=True,normalize=True,n_jobs=2)
lr.fit(X_train,y_tr)
```

```python
lr.score(X_train,y_tr)
```

```python
lr.score(X_test,y_ts)
```

```python
from sklearn.metrics  import mean_squared_error, mean_absolute_error ,r2_score
train_pred= lr.predict(X_train)
test_pred=lr.predict(X_test)

train_pred= np.expm1(train_pred)
test_pred=np.expm1(test_pred)
```

```python
sns.jointplot(train_pred,y_train)
```

```python
sns.jointplot(test_pred,y_test)
```

```python
plt.scatter(test_pred, test_pred- y_test, c='g', s=40, label='Residual Testing')
plt.scatter(train_pred, train_pred- y_train, c='b', s=40, label='Residual Testing')
```

```python
# Random forest
from sklearn.ensemble import RandomForestRegressor
from sklearn.cross_validation import train_test_split
from sklearn.metrics import make_scorer
from sklearn.metrics import mean_squared_error , r2_score

#xtrain, xtest, ytrain, ytest = train_test_split(x_data, y_data)



from sklearn.grid_search import GridSearchCV

# This helper function will make a callable that we can use in cross_val_score
mse_scorer = make_scorer(r2_score, greater_is_better=True)
#mse_scorer = make_scorer(mean_squared_error, greater_is_better=False)

#Parameter Optimisation with grid search
tuned_parameters = [{'max_features': ['sqrt', 'log2', 'auto'], 'max_depth': [15,25, 30, 40], 'min_samples_leaf': [ 5, 10, 15],'min_samples_split':[5,8,10]}]
rf =  GridSearchCV(RandomForestRegressor(n_jobs=2, n_estimators=50), tuned_parameters, cv=3, verbose=2, scoring=mse_scorer).fit(X_train,y_train )

```

```python
print ('Best Parameters:')
print (rf.best_estimator_)
```

```python
# Model Running with optimum parameters on entire training set
rf_opt = RandomForestRegressor(bootstrap=True, criterion='mse', max_depth=30,
           max_features='auto', max_leaf_nodes=None, min_samples_leaf=5,
           min_samples_split=10, min_weight_fraction_leaf=0.0,
           n_estimators=50, n_jobs=2, oob_score=False, random_state=None,
           verbose=0, warm_start=False)
rf_opt.fit(X_train, y_train)
```

```python

#Root Mean Squared Logarithmic Error (RMSLE)
print ("Root Mean Squared Error Train: ", np.sqrt(mean_squared_error(rf_opt.predict(X_train), y_train)))
print ("Root Mean Squared Error Test: ", np.sqrt(mean_squared_error(rf_opt.predict(X_test), y_test)))
print ("Training accuracy: %0.2f%%" % (rf_opt.score(X_train, y_train)))
print ("Test accuracy: %0.2f%%" % (rf_opt.score(X_test, y_test)) + "\n")
```

```python
plt.hist(target)
```

```python
plt.scatter(rf_opt.predict(X_test), y_test, c='g', s=40, label='Testing')
plt.scatter(rf_opt.predict(X_train), y_train, c='b', s=40, alpha = 0.25, label='Training')
plt.legend()
plt.axis('equal')
plt.title('Predicted Vs Real casual')
plt.xlabel("Predicted values")
plt.ylabel("Actual Values")
```

```python
plt.scatter(rf_opt.predict(X_test), rf_opt.predict(X_test)- y_test, c='g', s=40, label='Residual Testing')
plt.scatter(rf_opt.predict(X_train), rf_opt.predict(X_train)- y_train, c='b', s=40, alpha = 0.25, label='Residual Training')
plt.plot([-2,2],[0,0], c='r')
plt.legend()
plt.title('Residuals vs casual')
plt.xlabel("Predicted casual values")
plt.ylabel("Residuals")
```

```python
features=X_test.columns
numfeat = len(features)
indices = np.argsort(rf_opt.feature_importances_)[::-1][:numfeat]
plt.bar(xrange(numfeat), rf_opt.feature_importances_[indices], align='center', alpha=.5)
plt.xticks(xrange(numfeat), features[indices], rotation='vertical', fontsize=10)
plt.xlim([-1, numfeat])
plt.ylabel('Feature percentages', fontsize=12)
plt.title('Feature importances for demand by casual users', fontsize=18)
#print zip(features, rf_op.feature_importances_)
```
